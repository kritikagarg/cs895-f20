```python
import time

filename = "unique_accomplices_bidencheated.csv"
file1 = open(filename, 'r')
lines = file1.readlines()
file1.close()

#filename = "starcisco_followers.txt"
#file1 = open(filename, 'r')
#lines = file1.readlines()
#file1.close()


#for l in lines:
ratelimitCount = 0
totalCount = 1
currentPlace = 4568
f = open("botometer_scores.csv", "a")
#currentPlace = 1
#f = open("s_botometer_scores.csv", "a")
for l in lines:
    if totalCount >= currentPlace:
        try:
            overall = 0
            result = bom.check_account("@"+l.split('\n')[0])        
            if result['display_scores']['universal']['overall'] > result['display_scores']['english']['overall']:
                overall = result['display_scores']['universal']['overall']
            else:
                overall = result['display_scores']['english']['overall']
            print(l.split('\n')[0], str(overall))
            f.writelines(l.split('\n')[0] + ',' + str(overall) + '\n')
        except:
            print(l.split('\n')[0], "failed")            
        ratelimitCount += 1
        if ratelimitCount == 30:
            print("------- cool down -------")
            f.close()
            time.sleep(30)
            f = open("botometer_scores.csv", "a")
            #f = open("s_botometer_scores.csv", "a")
            ratelimitCount=0
    if totalCount == 6500:
        break
    totalCount += 1
    
len(lines)
f.close()
```
